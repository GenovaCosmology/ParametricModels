```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm

# Degrees of freedom range
d_values = np.arange(1, 1000)
# Q_obs = d + 25

Q_fixed = 25 + d_values

# Compute P(Q > Q_obs) = 1 - CDF(Q_obs)
p_values = chi2.sf(Q_fixed, df=d_values)

significance = norm.isf(p_values)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(d_values, significance,lw = 2, label=r'$P(Q > Q_{\mathrm{obs}})$', color='blue')
sigma_ticks = np.arange(1, 6)


plt.yticks(sigma_ticks, [fr'${s}\sigma$' for s in sigma_ticks])
# Find and mark the point closest to 2Ïƒ
plt.axhline(2,xmin = 0,xmax=0.108,linestyle = '--',color = 'k')
plt.axvline(68,ymin = 0,ymax=0.342,linestyle = '--',color = 'k')
xticks = plt.xticks()[0]  # get current xticks
new_xticks = np.append(xticks, 68)
plt.xticks(np.sort(new_xticks)[1:-1])

plt.xlabel(r'Degrees of Freedom ($d$)', fontsize=12)
plt.ylabel(r'$\mathrm{Significance}$', fontsize=12)
plt.legend(fontsize = 14)
plt.tight_layout()
plt.show()
```
