```python
import numpy as np
import pandas as pd
import scipy
from scipy.stats import chi2
from scipy.integrate import quad
from scipy.special import erfinv
import os
import getdist.plots as gdplt
from getdist import plots
from getdist import loadMCSamples

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
tempdir = os.path.join('../w0wa/chains/w0wa')
chain_path = os.path.join(tempdir, '04p1_w0wa_CMB_BAO_SN')
chain = loadMCSamples(chain_path,no_cache=True,settings={'ignore_rows':0.2})
chain2 = loadMCSamples(chain_path,no_cache=True,settings={'ignore_rows':0.2})
chain2.updateSettings({"contours": [0.99988]})
```

```python
g = plots.get_subplot_plotter(subplot_size=10)
g.settings.axes_fontsize=25
g.settings.legend_fontsize=25
g.settings.axes_labelsize=25
g.show_all_settings

# chain.updateSettings({"contours": [0.68,0.95,0.99988]})
g.settings.num_plot_contours = 3
g.plot_2d([chain], 'w','wa', filled=True,
          colors=[[16/256,139/256,150/256]],
          lims=[-1.1, -0.5, -1.5, 0.2])
g.settings.num_plot_contours = 1
g.add_2d_contours(chain2, 'w','wa', filled=False, color="#0B1D51", ls="--",)

g.add_x_marker(-1, linestyle='--', color='black',linewidth=1)
g.add_y_marker(0, linestyle='--', color='black',linewidth=1)
ax = g.get_axes()
ax.set_xlabel(r'$\theta_1$', fontsize=30)
ax.set_ylabel(r'$\theta_2$', fontsize=30)

# g.export('PSEfigure.pdf', dpi=600)
```
